```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Employee_Salaries_-_2023.csv')

# Preview the data
print(df.head())
```

	â€¢	What is the average salary for each department?

```python
avg_salary_by_job = df.groupby('Department_Name')['Base_Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_job)
```

How does gender effect salary?

```python
avg_salary_by_gender = df.groupby('Gender')['Base_Salary'].mean()
print(avg_salary_by_gender)

# Visualize salary distribution by gender
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Gender', y='Base_Salary', data=df)
plt.title('Salary Distribution by Gender')
plt.show()
```

 Is there a trend in salaries based on grade (as a proxy for experience)?

```python
df_grade_cleaned = df.dropna(subset=['Grade'])

# Group by grade and calculate average salary
avg_salary_by_grade = df_grade_cleaned.groupby('Grade')['Base_Salary'].mean().sort_index()
print(avg_salary_by_grade)

# Plot the trend
sns.lineplot(x='Grade', y='Base_Salary', data=df_grade_cleaned, estimator='mean')
plt.title('Salary Trend by Grade')
plt.show()
```

What are the top 5 highest-paying divisions?

```python
top_5_divisions = df.groupby('Division')['Base_Salary'].mean().sort_values(ascending=False).head(5)
print(top_5_divisions)
```

 Are salaries department-dependent?

```python
avg_salary_by_department = df.groupby('Department')['Base_Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_department)

# Plot the results
sns.barplot(x=avg_salary_by_department.index, y=avg_salary_by_department.values)
plt.title('Average Salary by Department')
plt.xticks(rotation=45)
plt.show()
```

Are there duplicates or missing data?

```python
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")

# Optionally, display the duplicate rows
if duplicate_rows > 0:
    duplicates = df[df.duplicated()]
    print("\nDuplicate rows:")
    print(duplicates)
```

What is the total overtime pay as a percentage of total base salary for each department?

```python
department_totals = df.groupby('Department_Name')[['Base_Salary', 'Overtime_Pay']].sum()

# Add a new column for overtime pay as a percentage of base salary
department_totals['Overtime_Percentage'] = (department_totals['Overtime_Pay'] / department_totals['Base_Salary']) * 100

# Sort by overtime percentage in descending order
overtime_percentage_sorted = department_totals.sort_values('Overtime_Percentage', ascending=False)
print("Overtime Pay as Percentage of Base Salary by Department:")
print(overtime_percentage_sorted)

# Optional: Visualize the results
import matplotlib.pyplot as plt

overtime_percentage_sorted['Overtime_Percentage'].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Overtime Pay as % of Base Salary by Department')
plt.ylabel('Overtime Percentage')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
